```python
class SVM:
    
    def __init__(self,C=1.0):
        self.C = C
        self.W = 0
        self.b = 0
        
    def hingeLoss(self,X,Y,W,b):
        m = X.shape[0]
        loss = 0.0
        
        loss += 0.5*np.dot(W,W.T)
        
        for i in range(m):
            ti = Y[i]*(np.dot(W,X[i].T)+b)
            loss += self.C *max(0,(1-ti))
        return loss[0][0]
    
    def fit(self,X,Y,max_itr=1000,batch_size=100,l_rate=0.0001):
        n_features = X.shape[1]
        n_samples = X.shape[0]
        
        c = self.C
        n = l_rate
        
        W = np.zeros((1,n_features))
        bias = 0
        
        ##calculationg inital loss
#         print(self.hingeLoss(X,Y,W,bias))
        
        #training data starts here
        losses = []
        
        for i in range(max_itr):
            l = self.hingeLoss(X,Y,W,bias)
            losses.append(l)
            ids = np.arange(0,n_samples)
            np.random.shuffle(ids)
        
        
        ##Batch Gradient descent with random shuffling
            for batch_start in range(0,n_samples,batch_size):
                gradw = 0
                gradb = 0
            
            #Iterate over all examples for mini batch
                for j in range(batch_start,batch_start+batch_size):
                    if j<n_samples:
                        i = ids[j]
                        ti = Y[i]*(np.dot(W,X[i].T)+bias)
                    
                        if ti>1:
                            gradw += 0
                            gradb += 0 
                        else:
                            gradw += c*Y[i]*X[i]
                            gradb += c*Y[i]
            
            #Gradient descent for the batch is ready! Update W and bias
                W = W - n*W + n*gradw
                bias = bias + n*gradb
                
        self.W = W
        self.b = bias
        return W,bias,losses
```

### Convert data for 1-vs-1 classification

```python
M = img_data.shape[0]
img_data = img_data.reshape((M,-1))
print(img_data.shape,labels.shape)
```

```python
CLASSES = len(np.unique(labels))
```

```python
def classWiseData(x,y):
    data = {}
    
    for i in range(CLASSES):
        data[i] = []
        
    for i in range(x.shape[0]):
        data[y[i]].append(x[i])
        
    for k in data.keys():
        data[k] = np.array(data[k])
        
    return data
```

```python
data = classWiseData(img_data,labels)
```

```python
data[0].shape[0],data[1].shape[0],data[2].shape[0],data[3].shape[0],
```

```python
def getDataPairForSVM(d1,d2):
    ## Combine data of two classes into single matrix
    l1 = d1.shape[0]
    l2 = d2.shape[0]
    
    samples = l1+l2
    features = d1.shape[1]
    
    data_pair = np.zeros((samples,features))
    data_labels = np.zeros((samples,))
    
    data_pair[:l1,:] = d1
    data_pair[l1:,:] = d2
    data_labels[:l1] = -1
    data_labels[l1:] = 1
    
    return data_pair,data_labels
```

## Training nC2 SVM

```python
mysvm = SVM()
```

```python
def trainSVMs():
    svm_classifiers = {}
    
    for i in range(CLASSES):
        svm_classifiers[i] = {}
        for j in range(i+1,CLASSES):
            xpair,ypair = getDataPairForSVM(data[i],data[j])
            wts,b,loss = mysvm.fit(xpair,ypair,max_itr=300)
            svm_classifiers[i][j] = (wts,b)
    return svm_classifiers
```

```python
svm_classifiers = trainSVMs()
```

### Prediction

```python
def binaryPredict(x,w,b):
    z = np.dot(x,w.T)+b
    if z >= 0:
        return 1
    else:
        return -1
```

```python
def predict(x):
    count = np.zeros((CLASSES,))
    
    for i in range(CLASSES):
        for j in range(i+1,CLASSES):
            w,b = svm_classifiers[i][j]
            #Take a majority prediction
            z = binaryPredict(x,w,b)
            if z==1:
                count[j] += 1
            else:
                count[i] += 1
                
    final_pred = np.argmax(count)
    return final_pred
```

```python
def accuracy(x,y):
    count = 0
    for i in range(x.shape[0]):
        
        pred = predict(x[i])
        if pred == y[i]:
            count+=1
#         print(pred)
    return  count/x.shape[0]
```
